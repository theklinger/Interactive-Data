```python
import numpy as np
import pandas as pd
from bokeh.models import HoverTool
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
twitch = pd.read_csv('twitchdata-update.csv')
twitch
```

```python
twitch_new = twitch.groupby('Language')
twitch_new.count()
```

```python
counts = twitch_new.count()
counts = counts.sort_values('Channel', ascending=False)
counts
```

```python
x = list(counts.index)
y = counts['Channel']

p = figure(x_range=x, 
           sizing_mode="stretch_width"
          )

p.line(x,y)

show(p)
```

```python

#Not part of the challenge:
x = list(counts.index)
y = counts['Channel']

#Add HoverTool Tip
hover1 = HoverTool(tooltips=[("Number of streamers", "@y")],
                  mode = 'vline'
                  )

p = figure(x_range=x, 
           sizing_mode="stretch_width", 
           height=300, 
           title="Twitch streamers by spoken Language",
           x_axis_label="Spoken Language", 
           y_axis_label="Number of Streamers", 
           toolbar_location=None, 
           tools=[hover1]
          )

p.line(x, y, line_color='green', legend_label="Number of streamers", line_width=2)


p.ygrid.grid_line_alpha = 0.8
p.ygrid.grid_line_dash = [6,4]

show(p)
```

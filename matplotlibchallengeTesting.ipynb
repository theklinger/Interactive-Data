```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('cces_example_data.csv')
```

```python
df_clean = df.astype({'sex': 'category', 'pid3': 'category', 'nonwhite': 'category', 'Evaluation': 'category'})
```

```python
fig, axs = plt.subplots()
axs.set_xlabel('Age Histogram')
axs.set_ylabel('Political Identity')
axs.set_title("Question #4")
```

```python
fig, ax = plt.subplots(figsize=(12,5))
categories = ['Evaluation']
ax.bar(df.index, categories)
ax.set_title("Question #5")
```
